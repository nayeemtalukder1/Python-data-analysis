```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=2)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=2, autopct='%1.2f%%')
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=2, autopct='%1.2f%%', shadow=True, explode=[0,0.2,0,0,0])
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals, labels=exp_labels,radius=2, autopct='%1.2f%%', shadow=True, explode=[0,0.2,0,0,0], startangle=45)
plt.show()
```
