```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,40,45,60,67,42]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'g+')
```

```python
plt.plot(x,y,'g+--')
```

<h3>Order does not important</h3>

```python
plt.plot(x,y,'g+-.')
```

```python
plt.plot(x,y,'--r*')
```

```python
plt.plot(x,y,'rD') #dimond
```

```python
plt.plot(x,y,color='red',marker='D',linestyle='')
```

```python
plt.plot(x,y,color='#000',marker='D',linestyle='') #use rgb color
```

```python
plt.plot(x,y,color='#000',marker='D',linestyle='',markersize=20) #use rgb color
```

```python
plt.plot(x,y,color='#000',marker='D',linestyle='',markersize=20,alpha=.2) #transparency
```
