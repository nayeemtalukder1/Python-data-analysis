```python
import pandas as pd
import openpyxl

def convert_people_cell(cell):
    if cell==44:
        return 'NAN'
    return cell
df = pd.read_excel("4.xlsx",sheet_name="3", converters = {
    'Age': convert_people_cell
})

df
```
